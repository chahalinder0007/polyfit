```python
import numpy as np,time
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"]=15,10
```

```python
x=[-20,-15,-12,-10,-5,-3,-2,0,2,3,5,10,11,15,20]
y=[0,-150,-220,-210,-100,0,0,0,0,0,100,200,210,150,0]

coef=np.polyfit(x,y,deg=8)
xx=np.linspace(-20,20,num=25)
```

```python
def func(xx):
    
    yy=0
    degre=len(coef)-1
    for i in range(len(coef)):
        c=coef[i]
        
        yy+=c*(xx**degre)
        degre-=1
        
    #yy=(coef[0])*(xx)**6 +(coef[1])*(xx)**5 + (coef[2])*(xx)**4 + (coef[3])*(xx)**3 + (coef[4])*(xx)**2 + (coef[5]*xx)+coef[6]
    return yy

yy=[func(g) for g in xx]
plt.plot(xx,yy)
plt.plot(x,y)
plt.savefig("dabba.png")
```
